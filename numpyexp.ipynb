```python
import numpy as np
arr=np.array([10,20,30,40,50])
mean_val=np.mean(arr)
std_dev=np.std(arr)
max_val=np.max(arr)
min_val=np.min(arr)
print(arr)
print("The Average is",mean_val)
print(f"The Standard Deviation is {std_dev:.2f}")
print("The Maximum Value is",max_val)
print("The Minimum Value is",min_val)
```

```python
import numpy as np
A=np.array([[1,2],[3,4]])
B=np.array([[5,6],[7,8]])
C=np.dot(A,B)
print("Matrix A:\n",A)
print("Matrix B:\n",B)
print("Matrix Multiplication (A X B):\n",C)
```
